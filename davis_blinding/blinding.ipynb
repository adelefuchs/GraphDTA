#### 20% blinding

```python
import pandas as pd
import random

# Load the datasets
train_df = pd.read_csv("../data/davis_train.csv")
test_df = pd.read_csv("../data/davis_test.csv")
other20 = pd.read_csv("data/test20.csv")

# Get proteins
train_proteins = set(train_df["target_sequence"])
test_proteins = set(test_df["target_sequence"])
other20_proteins = set(other20["Target Sequence"])

# Combine proteins from train and test datasets
all_proteins = train_proteins.union(test_proteins)

# Find intersection with other20 proteins
overlapping_proteins = other20_proteins.intersection(all_proteins)

# Calculate the percentage of overlapping proteins
percentage = (len(overlapping_proteins) / len(all_proteins)) * 100
print(f"Proteins in other20 make up {percentage:.2f}% of the combined train and test datasets.")

# Combine train and test datasets
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# Separate proteins to blind from the combined set
proteins_to_blind = other20_proteins

# Find rows in combined_df where target_sequence is in proteins_to_blind
blinded_interactions = combined_df[combined_df["target_sequence"].isin(proteins_to_blind)]

# Find non-blinded interactions
non_blinded_interactions = combined_df[~combined_df["target_sequence"].isin(proteins_to_blind)]

# Shuffle the non-blinded interactions
non_blinded_list = non_blinded_interactions.to_dict(orient="records")
random.shuffle(non_blinded_list)

# Convert the shuffled non-blinded data back to DataFrame
train_df_blinded = pd.DataFrame(non_blinded_list)

# The test set should consist of the blinded proteins
test_df_blinded = blinded_interactions

# Save the updated datasets
train_df_blinded.to_csv("data/davis_b3_train.csv", index=False)
test_df_blinded.to_csv("data/davis_b3_test.csv", index=False)

print("Datasets with blinded proteins and non-blinded data have been saved to 'train_b3.csv' and 'test_b3.csv'.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the CSV file
df = pd.read_csv('predictions_GINConvNet_davis_b.csv')

# Extract true and predicted values
y_true = df['True_Label'].values
y_pred = df['Predicted_Value'].values


# Compute regression metrics
mse = mean_squared_error(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)
pearson_corr, _ = stats.pearsonr(y_true, y_pred)
spearman_corr, _ = stats.spearmanr(y_true, y_pred)

# Print metrics
print(f"Mean Squared Error (MSE): {mse:.3f}")
print(f"Mean Absolute Error (MAE): {mae:.3f}")
print(f"RÂ² Score: {r2:.3f}")
print(f"Spearman Correlation: {spearman_corr:.3f}")
print(f"Pearson Correlation: {pearson_corr:.3f}")

# Scatter plot of true vs. predicted values
plt.figure(figsize=(6, 6))
plt.scatter(y_true, y_pred, alpha=0.5, color='blue', label='Predictions')
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], color='red', linestyle="--", label='Ideal Fit')  # Identity line
plt.xlabel("True Affinity")
plt.ylabel("Predicted Affinity")
plt.title("True vs. Predicted Affinities")
plt.legend()
plt.show()
```
